## Matplotlib Operation used for visualizations

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline 
```

```python
#To plot 2D graph I need some data I am using array 
x=np.arange(0,10)
y=np.arange(12,22)
```

```python
#making scatter graph

plt.scatter(x,y,c='g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D- Graph')
#plt.savefig('2DGraph.png')
```

```python
y=x*x
```

```python
#just by using plot
plt.plot(x,y,'g*',linestyle='dashed',linewidth=2, markersize=12)
#plt.plot(x, y, color='green', marker='*', linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D- Graph using plot')
```

#### Subplots


```python
## creating subplot

## plt.subplot(nrow,ncols,position)
plt.subplot(2,2,1)
plt.plot(x,y,'r-')
plt.subplot(2,2,2)
plt.plot(x,y,'g*')
plt.subplot(2,2,3)
plt.plot(x,y,'b*-')
plt.subplot(2,2,4)
plt.plot(x,y,'mo')
```

#### Sin & cos Graph

```python
#sin graph
x=np.arange(0,4* np.pi,0.1)
y=np.sin(x)
plt.title("sin - Graph")
plt.plot(x,y)
```

```python
#cos graph
x=np.arange(0,4* np.pi,0.1)
y=np.cos(x)
plt.title("cos - Graph")
plt.plot(x,y,'r')
```

#### Bar plot

```python
## Bar plot

x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')
```

#### Histogram

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

#### Boxplot

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);
```

#### Piechart


```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
